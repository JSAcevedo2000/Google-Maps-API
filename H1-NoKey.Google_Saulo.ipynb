A. Use the [*Nearby Seach*](https://developers.google.com/places/web-service/search) Method to find all the restaurant near to the UPY (20.988459, -89.736768) around 2km. In order to fulfill this item you need to:
* Create a *search_nearby_places* function with the appropiate parameters. It must return a JSON with the results. Places must be ranked by distance.
* Create a table with the following columns: Name, Place Id, Rating, Place Types, Total of User Ratings.

```python
import pandas as pd
import json
import requests
import time
```

```python
"""Find and return all the nearby restaurants from a given place.

The following function uses the Google Maps Places API to obtain and return all the restaurants that are in a given range, from a given point.
This function could also look for other than restaurants, however that would require to modify the API link (type = kind of places).

  Typical usage example:

  search_nearby_places(latitude, longitude, radius of search, API key), where:
      -Latitude and longitude refer to the ubication of the place we are taking as base. It is possible to find them from the Google Maps App (INTEGERS).
      -Radius of search is the radius of the circle in which we are looking for Restaurants, with origin at the place we select (INTEGER).
      -API key is our Google Maps Cloud Services key, it is not required to pay for it but you need to register in order to obtain it (STRING).
"""

def search_nearby_places(lat, lon, radius, key):
    data = json.loads(requests.get('https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={},{}&radius={}&key={}&type=restaurant'.format(lat, lon, radius, key)).content)
    df = pd.DataFrame(data['results'])
    df = df.drop(['geometry', 'icon', 'id', 'opening_hours', 'photos', 'plus_code', 'reference', 'scope', 'vicinity'], axis = 1)
    return df

search_nearby_places(20.988459, -89.736768, 2000, 'API KEY HERE').head()
```

B. Create a function to Find the Place Id given Place Name (String) and Location (Lat/Lont Tuple).

```python
"""Find the Google Maps ID of a given place.

This function uses the location data (name of the place, latitude and longitude) to find its ID.
The ID of the place is the unique number of identification provided by Google Maps.

  Typical usage example:

  placefinder(name of the place, location, API key), where:
      -Name of the place is literally the name. Quotation marks should be deleted. (STRING)
      -The location must be written as a tuple of latitude and longitude, for example: (20.999, -80.999). (INTEGERS - TUPLE)
      -API key is our Google Maps Cloud Services key, it is not required to pay for it but you need to register in order to obtain it (STRING).
"""

def placefinder(place, location, key):
    placedata = json.loads(requests.get('https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={}&inputtype=textquery&locationbias={},{}&key={}'.format(place, location[0], location[1], key)).content)
    placeid = placedata['candidates'][0]['place_id']
    return placeid 
```

D. Use the [*Place Details*](https://developers.google.com/places/web-service/details#PlaceDetailsResults) Method to find the Reviews of a Place in English and Spanish. In order to fulfill this item you need to:
* Create a *find_reviews* function with the appropiate parameters. It must return a JSON with the results.

```python
"""Find Google Maps reviews from a given place.

This function uses the ID from a given place (which can be obtained with the previous function) and our prefered language to find reviews.

  Typical usage example:

  find_reviews(place ID, language, API key), where:
      -Place ID refers to the Google Maps identification number of the place we are looking for (STRING).
      -Language refers to the abbreviation of the reviews' language we are looking for (English = en, Spanish = es) (STRING, separeted by commas).
      -API key is our Google Maps Cloud Services key, it is not required to pay for it but you need to register in order to obtain it (STRING).
"""

def find_reviews(placeid, language, key):
    reviewsdata = json.loads(requests.get('https://maps.googleapis.com/maps/api/place/details/json?place_id={}&language={}&key={}'.format(placeid, language, key)).content)
    placereviews = pd.DataFrame(reviewsdata['result']['reviews'])
    placereviews = placereviews.drop(['author_url', 'profile_photo_url', 'relative_time_description'], axis = 1)
    return placereviews
```

```python
#D. Use the [*Place Details*] Method to find the Reviews of a Place in English and Spanish. In order to fulfill this item you need to:
#* Create a *find_reviews* function with the appropiate parameters. It must return a JSON with the results. 
#* Create a table with the following columns: Author Name, Language, Rating, Text, Time.

"""Find Google Maps reviews from a given place.

This function uses the ID from a given place (which can be obtained with the previous function) to find reviews in English and Spanish.

  Typical usage example:

  find_reviews(place ID), where:
      -Place ID refers to the Google Maps identification number of the place we are looking for (STRING).
"""

import pandas as pd
import json
def find_reviews(placeid):
    reviews = maps.place(place_id = placeid, language='en,es')
    reviews = pd.read_json(json.dumps(reviews['result']['reviews']), orient='records')
    reviews = reviews.drop(['author_url', 'profile_photo_url', 'relative_time_description'], axis = 1)
    return reviews
find_reviews('ChIJaUUfylFxVo8RgEGjoansPbc')
```

```python
#E. Find similar places to Los Trompos Circuito. Rank them by its "Prominence". In order to fulfill this item you need to:
#* Create a *find_similar* function with the appropiate parameters. It must return a JSON with the results.
#* Create a table with the following columns: Place Name, Plaece Id, Rating, Total of Use Ratings.

"""Find places similar to a given place.

This function uses the location of a given place in order to find similar places, based on the "Prominence" or the popularity of the places.

  Typical usage example:

  find_similar((latitude, longitude)), where:
      -Latitude and Longitude both refer to the location of the place. The info can be obtained from the Google Maps App (2 Integers Tuple - STRING).
"""

def find_similar(locatn):
    similar = maps.places_nearby(location = locatn, radius=1000, type='restaurant', rank_by='prominence')
    similar = pd.read_json(json.dumps(similar['results']), orient='records')
    similar = similar[['name', 'id', 'rating', 'user_ratings_total']]
    return similar
find_similar('21.000130, -89.617245').head()
```
